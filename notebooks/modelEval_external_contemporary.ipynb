```python
from vecanalysis import alignment
from representations.representation_factory import create_representation
from ioutils import write_pickle, words_above_count, mkdir

def align_years(years, rep_type, in_dir, out_dir, words, **rep_args):
    first_iter = True
    base_embed = None
    for year in years:
        print("Loading year:", year)
        year_embed = create_representation(rep_type, in_dir + str(year), **rep_args)
        year_words = words[str(year)]
        year_embed.get_subembed(year_words)
        print("Aligning year:", year)
        if first_iter:
            aligned_embed = year_embed
            first_iter = False
        else:
            aligned_embed = alignment.smart_procrustes_align(base_embed, year_embed)
        base_embed = aligned_embed
        print("Writing year:", year)
        foutname = out_dir + str(year)
        np.save(foutname + "-w.npy",aligned_embed.m)
        write_pickle(aligned_embed.iw, foutname + "-vocab.pkl")
        
align_years([12012, 10000], 'word2vec', BASEPATH_EMBEDDINGS, BASEPATH_EMBEDDINGS + "aligned_", {'12012': list(df_embeds2012['word']), '10000': list(df_wikipedia2vec['word'])})
```

```python
aligned_wikipedia_embeds = np.load(open("resources/word_embeddings/aligned_10000-w.npy",  "rb"))
with open(os.path.join("resources/word_embeddings/aligned_10000-vocab.pkl"), 'rb') as f:
    aligned_wikipedia_vocab = pickle.load(f)

df_aligned_wikpedia2vec = pd.DataFrame(aligned_wikipedia_embeds)
df_aligned_wikpedia2vec["word"] = aligned_wikipedia_vocab
```

```python
with open("regression_model_trainset.pkl", 'rb') as f:
    model = pickle.load(f)
with open("testset_words.pkl", 'rb') as f:
    testwords = pickle.load(f)

df_aligned_wikpedia2vec = df_aligned_wikpedia2vec[df_aligned_wikpedia2vec['word'].isin(testwords)]
df_aligned_wikpedia2vec['valence_pred'] = model.predict(df_aligned_wikpedia2vec.iloc[:, :-1])

df_aligned_wikpedia2vec = pd.merge(df_aligned_wikpedia2vec, df_xanew, on='word', how='inner')
```

```python
msre = sklearn.metrics.mean_squared_error(df_aligned_wikpedia2vec['valence_gold'], df_aligned_wikpedia2vec['valence_pred'])
print("MSRE of model: " + str(msre))
msreDummy = sklearn.metrics.mean_squared_error(df_aligned_wikpedia2vec['valence_gold'], [5.0] * len(df_aligned_wikpedia2vec['valence_pred']))
print("MSRE of dummy model: " + str(msreDummy))
```
